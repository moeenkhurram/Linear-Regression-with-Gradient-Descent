# Linear Regression with Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
# y = mx + b
# m is slope, b is y-intercept

class LinearRegression_with_GradientDescent:
    
    def __init__(self, learning_rate=0.01, iterations=100):
        self.learning_rate, self.iterations = learning_rate,  iterations
        
    #method fit
    def fit(self, X, y):
        b_current = 0
        m_current = 0
        n = X.shape[0]
        
        for _ in range(self.iterations):
            y_current = (m_current * X) + b_current
            cost = sum(((y-y_current)**2) / n)
            
            b_gradient = -2 * np.sum(y - y_current) / n
            m_gradient = -2 * np.sum(X*(y - (y_current))) / n
            
            b_current = b_current - (self.learning_rate * b_gradient)
            m_current = m_current - (self.learning_rate * m_gradient)
            
        self.m_current, self.b_current, self.cost = m_current, b_current, cost
        
    def predict(self, X):
        return self.m_current*X + self.b_current
```


# Data

```python
np.random.seed(786)
X = np.array(sorted(list(range(5))*160)) + np.random.normal(size=800, scale=0.5)
y = np.array(sorted(list(range(5))*160)) + np.random.normal(size=800, scale=0.25)
```

```python
clf = LinearRegression_with_GradientDescent(learning_rate=0.01,iterations=1000)
clf.fit(X, y)


plt.scatter(X, y, color='red')
plt.plot(X, clf.predict(X))
plt.gca().set_title("Linear Regressor with Gradient Descent ")
```

```python
print('Cost:',clf.cost)
print('slop (m):',clf.m_current)
print('Intercept (b):',clf.b_current)
```
